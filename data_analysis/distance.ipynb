```python
class JudgeRegion():
    """
    used to detect region for sid
    """
    def getPara(self, num):
        y = num % 100
        floor = int(num / 10000)
        x = int((num - floor * 10000) / 100)

        return x, y, floor

    def getWhere(self, num):
        if num == 0:
            return 'out'

        x, y, floor = self.getPara(num)

        if floor == 1:
            if x == 15 and (y == 5 or y == 15 or y == 17) or x == 0 and y == 19:
                return 'out'

            if 2 <= x < 4 and 1 <= y < 6:
                return 'hallA'

            if 4 <= x < 6 and 1 <= y < 6:
                return 'hallB'

            if 6 <= x < 8 and 1 <= y < 6:
                return 'hallC'

            if 8 <= x < 10 and 1 <= y < 6:
                return 'hallD'

            if 3 <= x < 10 and 7 <= y < 9:
                # return 'other'
                return 'poster'

            if 4 <= x < 6 and 10 <= y < 12:
                return 'toilet1'
                # return 'other'

            if 6 <= x < 10 and 10 <= y < 12:
                return 'room1'
                # return 'other'

            if 10 <= x < 12 and 10 <= y < 12:
                return 'room2'
                # return 'other'

            if 2 <= x < 12 and 15 <= y < 19:
                return 'exhibition'

            if 2 <= x < 12 and 19 <= y < 29:
                return 'mainHall'

            if 12 <= x < 14 and 2 <= y < 6:
                return 'register'
                # return 'other'

            if 14 <= x < 16 and 19 <= y < 21:
                return 'service'
                # return 'other'

            if 14 <= x < 16 and 21 <= y < 25:
                # return 'other'
                return 'room3'

            if 14 <= x < 16 and 25 <= y < 27:
                # return 'other'
                return 'room4'

            if 14 <= x < 16 and 27 <= y < 29:
                # return 'other'
                return 'toilet2'

            if x == 1 and 10 <= y < 12:
                # return 'Up'
                return 'elevatorUp'
                # return 'other'

            if x == 14 and 10 <= y < 12:
                # return 'Down'
                return 'elevator'
                # return 'other'

            return 'other'

        else:
            if 2 <= x < 10 and 1 <= y < 6:
                # return 'other'
                return 'diningHall'

            if 10 <= x < 12 and 1 <= y < 6:
                # return 'other'
                return 'room5'

            if 13 <= x and 0 <= y < 6:
                # return 'other'
                return 'recreationArea'

            if 4 <= x < 6 and 10 <= y < 12:
                # return 'other'
                return 'toilet3'

            if 6 <= x < 8 and 10 <= y < 12:
                # return 'other'
                return 'room6'

            if x == 1 and 10 <= y < 12:
                # return 'other'
                return 'elevatorUp'

            if x == 14 and 10 <= y < 12:
                # return 'other'
                return 'elevator'

            return 'other'
        
```

```python
dictSaved = []
region = JudgeRegion()
for i in range(3):
    person_room_count = {}
    day_data = pd.read_csv('data/day{}.csv'.format(i+1))
    
    day_data = day_data.sort_values(by=['id', 'time'], axis=0)
    
    last_person_id = 0
    for index, row in day_data.iterrows():
        if row['id'] == last_person_id:
            r = region.getWhere(row['sid'])
            if r in person_room_count[last_person_id]:
                pass
            else:
                person_room_count[last_person_id].append(r)
        else:
            last_person_id = row['id']
            person_room_count[last_person_id] = []
            person_room_count[last_person_id].append('other')
    dictSaved.append(person_room_count)

ss = []
for i in range(3):
    s = {}
    for key, value in dictSaved[i].items():
        s[key] = len(value)
    ss.append(s)
    print(s)
```

```python
list_total = []
for i in range(3):
    list_total.append([])
    tdf = pd.read_csv('feature_set{}.csv'.format(i+1))
    
    for key, value in ss[i].items():
        list_total[i].append(value)
    
    tdf.insert(tdf.shape[1], 'Rooms Count', list_total[i], True)
    
    tdf.to_csv('feature_set{}_0604.csv'.format(i+1))
        
```
