```python
import pandas as pd

data = pd.read_csv('feature_all.csv')

print(data.columns)
```

row 不管用

```python
T_rec = 0
in_max = 0
out_min = 1e6
Room_count_rec = 0
for index, row in data.iterrows():
    data.iloc[index, 2] = 1 / data.iloc[index, 2]
    if data.iloc[index, 2] > T_rec:
        T_rec = data.iloc[index, 2]
    if row['Rooms Count'] > Room_count_rec:
        Room_count_rec = row['Rooms Count']
    if row['in'] > in_max:
        in_max = row['in']
    if row['out'] < out_min:
        out_min = row['out']
print(T_rec)
print(in_max)
print(out_min)
print(Room_count_rec)
```

```python
thres_in = 28500
thres_out = 63300
thres_T = T_rec
thres_room_count = Room_count_rec
thres_stime = 200

for index, row in data.iterrows():
    data.iloc[index, 2] = row['T'] / thres_T * 0.8
    data.iloc[index, 9] = row['Rooms Count'] / thres_room_count * 0.8
    data.iloc[index, 3] = ( in_max - row['in'])/ (in_max - thres_in) * 0.8
    data.iloc[index, 4] = (row['out'] - out_min)/(thres_out - out_min)* 0.8 
    data.iloc[index, 6] = row['Stay Time'] * 0.8 / thres_stime
print(data)
```

```python
data.to_csv('regulization1.csv')
```
