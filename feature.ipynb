```python
import pandas as pd
import json
import numpy as np
```

```python
files = [
    'regulization/regulization1.csv',
    'regulization/regulization2.csv',
    'regulization/regulization3.csv'
]
```

```python
df_day1 = pd.read_csv(files[0])
df_day2 = pd.read_csv(files[1])
df_day3 = pd.read_csv(files[2])
df = pd.DataFrame(columns=['id', 'Room', 'Stay Time', 'Been Room5', 'Been Room6', 'Rooms Count', 'Been Room4', 'No Register'])
id = []
for i in range(df_day1.__len__()):
    if df_day1['id'][i] in id:
        continue
    else:
        id.append(df_day1['id'][i])
for i in range(df_day2.__len__()):
    if df_day2['id'][i] in id:
        continue
    else:
        id.append(df_day2['id'][i])
for i in range(df_day3.__len__()):
    if df_day3['id'][i] in id:
        continue
    else:
        id.append(df_day3['id'][i])
print(id)
print(df_day1.columns[6])
```

```python
for i in range(len(id)):
    day1 = df_day1[df_day1['id'] == id[i]]
    day1.index = np.arange(day1.__len__())
    day2 = df_day2[df_day2['id'] == id[i]]
    day2.index = np.arange(day2.__len__())
    day3 = df_day3[df_day3['id'] == id[i]]
    day3.index = np.arange(day3.__len__())
    frames = [day1, day2, day3]
    result = pd.concat(frames)
    result.index = np.arange(result.__len__())       
    
    stay_time = result['Stay Time'].max(axis=0)
    been_room5 = result['Been Room5'].max(axis=0)
    been_room6 = result['Been Room6'].max(axis=0)
    rooms_count = result['Rooms Count'].max(axis=0)
    been_room4 = result['Been Room4'].max(axis=0)
    no_register = result['No Register'].max(axis=0)
    
    room = result['Room'][np.argmax(result['Stay Time'],axis=0)]
        
    new=pd.DataFrame({'id':id[i],
                  'Room':room,
                'Stay Time':stay_time,
                 'Been Room5':been_room5,
                  'Been Room6':been_room6,
                  'Rooms Count':rooms_count,
                   'Been Room4':been_room4,
                   'No Register':no_register},index=[1])
    df=df.append(new,ignore_index=True)
df = df.sort_values(by=["id"], axis=0)
df.index = np.arange(df.__len__())
print(df)
df.to_csv('feature.csv')
```
